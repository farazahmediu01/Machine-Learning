```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Plots from krish naik tutorial
a = np.arange(0,10)
b = np.arange(10,20)
x = np.arange(40,50)
y = np.arange(50,60)

# Creating multiple plots or subplots in one diagram.
plt.subplot(2,2,1) # plt.subplot(*numRows*, *numCols*, *plotNum*) 
plt.plot(x,y,c='g')
plt.subplot(2,2,2) # 2 raws 2 columns and 2nd position.and
plt.plot(y,y**y,c='m')
plt.subplot(2,2,4) # this plot will go to 4th position
plt.plot(y*(-a),x,c='b')
```

```python
squares = [x**2 for x in range(1,6)]
squares
```

### Plotting a Simple Line Graph
- The function plt.subplots() can generate one or more plots in the same fig­ure. The variable fig represents the entire figure or collection of plots that are generated. The variable ax represents a single plot in the figure and is the variable we’ll use most of the time.

```python
fig, ax = plt.subplots()
ax.plot(squares)
```

### Changing the Label Type and Line Thickness
Matplotlib allows you to adjust every feature of a visualization.
We’ll use a few of the available customizations to improve this plot’s
readability, as shown here:

```python
fig, ax = plt.subplots()
ax.plot(squares, color='b', linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# # Set size of tick labels.
# ax.tick_params(axis='both', labelsize=14)
```

```python
fig, ax = plt.subplots()
ax.plot(squares, 'g-o', linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# # Set size of tick labels.
ax.tick_params(axis='both', labelsize=12)
```

### Correcting the Plot
- But now that we can read the chart better, we see that the data is not plot­ted correctly. Notice at the end of the graph that the square of 4.0 is shownas 25! Let’s fix that.
- When you give plot() a sequence of numbers, it assumes the first data point corresponds to an x­coordinate value of 0, but our first point corre­ sponds to an x­value of 1. We can override the default behavior by giving plot() the input and output values used to calculate the squares:


```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()
ax.plot(input_values, squares, 'g-o', linewidth=3)

# Set chart title and label axes.
ax.set_title("Square Numbers", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

# # Set size of tick labels.
ax.tick_params(axis='both', labelsize=12)
```

### Plotting and Styling Individual Points with scatter()
