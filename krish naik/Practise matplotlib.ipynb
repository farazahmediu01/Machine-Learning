```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
squares = [x**2 for x in np.arange(1,6)]
squares
```

```python
plt.plot(squares)
```

```python
fig, ax = plt.subplots()
ax.plot(squares)

plt.title("Square of Numbers.")
plt.xlabel("Value")
plt.ylabel("Square of Value")
```

```python


fig, ax = plt.subplots()
ax.plot(squares, c='b', linewidth=2)

ax.set_title("Square of Numbers.", fontsize=24)
ax.set_xlabel("Value", fontsize=14)
ax.set_ylabel("Square of Value", fontsize=14)

ax.tick_params(axis='both', labelsize=12)
```
