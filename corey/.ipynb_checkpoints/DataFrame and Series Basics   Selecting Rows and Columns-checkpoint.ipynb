# Introduction to DataFrame and DataSeries
### Definitions
<u><b>DataFrame</b></u>
- DataFrames are two dimentional data structues which contains data in a form of raws and columns and their corrosponding lables.
- DataFrames are similar to python dictionaries but with more functionality than what we have in native python.

<b><u>Data Series</u></b>
- DataSeries are usually the raws or the columns of a DataFrame with labels. For a raw_series lables are column names and for a column_series lables are raw indexes.
- It say that Series is a One dimensional array.

### Topics Covered
- Introducing DataFrames.
- Basic DataFrame functionality.
- Accessing column/columns.
- Introducing Series.
- Accessing columns with <b>df\["column"\]</b> and <b>df.column</b> and why we should avoid using dot notation for accessing column.
- Accesing specific and multiple columns with <b>df.\"columns1", "column3"\]</b>
- Get all columns name with <b>df.columns</b>
- Accessing specific raws and columns in specifci order with <b>iloc[]</b> and <b>loc[]</b>.
- Working with stackoverflow data.
- Counting how many says "Yes" and how many says "No" in almost 70,000 records using <b>df.value_counts()</b>
