# How to Set, Reset, and Use Indexes

# Topics Covered
- Setting a Column as a Raw_Index of a DataFrame using <b>df.set_index("column_name", inplace=True)</b>.
- See indexs with <b>df.index</b>
- Reset previous indexs with <b>df.reset_index(inplace=True)</b>
- Read full text of a block by selecting that block using <b>df.loc\["raw"\]\[index\] or df.loc\["raw","column"\] or df.at\["index","column"\]</b>
- Sort DataFrame by index with <b>df.sort_index(inplce=optionl)</b>
- Sort DataFrame by index in decending order <b>df.sort(ascending=False)</b>
