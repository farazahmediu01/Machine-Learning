### Topics we cover in this Tutorial
- Plotting a scatter plot with <b>plt.scatter(data,data,color)</b>

- When we have to display a Matplotlib graph or diagrams we have to write plt.show()
- %matplotlib inline # By this line of code we do not have to write plt.show() every time

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Simple Example
a = np.arange(0,10)
b = np.arange(10,20)
x = np.arange(40,50)
y = np.arange(50,60)
```

```python
plt.plot(a,b)
```

```python
# Plotting a scatter graph using matplotlib.
plt.scatter(a,b,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
print(x,y)
print(a,b)
```

```python
# Plot with plt
plt.plot(a,a*a) 
```

```python
# PLot with plt whith changing color of dotted line.
plt.plot(a,a*a,'m-o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
# Creating multiple plots or subplots in one diagram.plt

plt.subplot(2,2,1) # plt.subplot(raw, column, index)
plt.plot(x,y,c='g')
plt.subplot(2,2,2) # 2 raws 2 columns and 2nd position.and
plt.plot(y,y**y,c='m')
plt.subplot(2,2,4) # this plot will go to 4th position
plt.plot(y*(-a),x,c='b')
```

```python

plt.title('Matplotlib Demo')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(y*y,a*(-a),'m-o')
```

```python
# Compute the x and y co-ordinates for points on a sine curve
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
# Ploting the points using matplotlib
plt.title('sine wave form')
plt.plot(x,y)
```

```python
# Bar Plot
x1 = [2,8,10]
y1 = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1)
plt.bar(x2,y2,color='g')
plt.title('Bar-Graph')
plt.xlabel('X-axis')  # x values
plt.ylabel('Y-axis')  # y values
```

```python
# Histogram
plt.hist(x1+y1) # values exits on x-axis value or count appers on y-axis
plt.title('Histogram')
```
