# Pandas Tutorial 02

### Topics we cover in this Tutorial
Following operation we will cover in this Pandas tutorial.
1. <b>df.info()</b> This method prints information about a DataFrame including the index dtype and column dtypes, non-null values and memory usage. Inshort it print a concise summary of a DataFrame.
2. <b>df.describe()</b> It provides generate descriptive statistics.
3. <b>DataSeries.value_counts()</b> This will tell us how many times an object is presented in a column or in a raw.
4. Filter Data from columns.
5. Converting a strings into a csv file with <b>StringIO</b> then reading it with <b>pd.read_csv()</b>.
and introducing some key_word arguments of pandas read csv method like URL to csv an following. 
    - <b>pd.read_csv(filepath_or_buffer, index_col, usecols, dtype, sep, escapechar)</b>.
6. Checking data types of all the columns with <b>df.dtypes</b> attribute.
7. cheching memory usgae with <b>df.memory_usage()</b> method.

```python
# 5
from io import StringIO
```

```python
# 5
data = (",c1,c2,c3,c4,c5\n" # removed coma before c1.
        "101,1,2,3,4,5\n"
        "102,6,7,8,9,10\n"
        "103,11,12,13,14,15\n"
        "104,16,17,18,19,20\n")
str_io = StringIO(data, newline='\n') # we are converting a string into a csv file.
```

```python
# 5 
csv = pd.read_csv(StringIO(data),
                  usecols=['c2', 'c3', 'c4', 'c5'],
                  dtype={ 'c2': 'int32', 'c3': 'int', 'c4':'float32'})    # reading csv with only specifeid columns.
csv
```

```python
csv.dtypes # 6
```

```python
csv.memory_usage() # 7
```
